```python
#Lab Session
#Case Study Activity: UMaT Management System
class Person:
    def __init__(self,name,age):
        self.name = name
        self.age = age

    def display_info(self):
        print(f"Name:{self.name}, Age:{self.age}")

class Student(Person):
    def __init__(self,name,age,student_id,enroll_course):
        super().__init__(name,age)
        self.student_id = student_id
        self.enroll_course = enroll_course

    def display_info(self):
        print(f"I am {self.name} \nI am {self.age} \nI have Student ID {self.student_id} \nMy Course Enrollment Status = {self.enroll_course}")

class Lecturer(Person):
    def __init__(self,name,age,employee_id,teach_course):
        super().__init__(name,age)
        self.employee_id = employee_id
        self.teach_course = teach_course

    def show_more_info(self):
        print(f"HE is a Lecturer with Employee_id: {self.employee_id} and Course Teaching Status: {self.teach_course} ")


student = Student("Woode",19,9016915625,True)
student.display_info()

lecturer = Lecturer("Dr.Matthew Cobbinah",30,"1234567",True)
lecturer.display_info()
lecturer.show_more_info()
```
